# 1. CartPole simulation
카트 위에 놓인 막대가 넘어지지 않도록 카트를 왼쪽이나 오른쪽으로 가속할 수 있는 2D 시뮬레이션  
어떤 행동을 취하든 초당 1의 보상을 받는 이 시스템의 목적은 막대를 넘어트리지 않고 최대한 오랫동안 실행하는 것임.  

```python
import gym 

env = gym.make("CartPole-v1") # 원하는 시뮬레이션 환경설정
obs = env.reset()             # 환경 초기화
print(obs)                    # 카트의 수평위치, 카트의 속도, 막대 각도, 각속도
```

```python
env.render()
```

```python
def plot_environment(env, figsize=(5,4)):
    plt.figure(figsize=figsize)
    img = env.render(mode="rgb_array")
    plt.imshow(img)
    plt.axis("off")
    return img

# can get ndarray type image
img = env.render(mode = "rgb_array")

plot_environment(env)
plt.show()
```

```python
print(env.action_space)   # possible action - Discrete(2) : 취할수있는 행동, 왼쪽(0)과 오른쪽(1) 가속 
```

```python
action = 1  # accelerate right
obs, reward, done, info = env.step(action)
print(obs)
print(reward)
print(done)
print(info)

res = plot_environment(env)
```

step 메서드는 행동을 input하여 실행하고 4가지 값을 반환한다.
1. obs : 행동 실행 후 새로운 관측값
2. reward : 행동 실행 후 받는 보상
3. done : True or False의 값을 갖는데, True이면 에피소드(한 판)가 끝난 것
4. info : 디버깅이나 훈련에 유용한 추가적인 정보가 담긴다. (게임에서 남은 life 등)

### 1.1 simple policy hard coding

```python
# 간단한 정책을 하드코딩
def basic_policy(obs):
    angle = obs[2]
    return 0 if angle < 0 else 1

totals = []
for episode in range(500):
    episode_rewards = 0
    obs = env.reset()
    for step in range(200):
        action = basic_policy(obs)
        obs, reward, done, info = env.step(action)
        episode_rewards += reward
        if done:
            break
    totals.append(episode_rewards)
```

```python
import numpy as np
np.mean(totals), np.std(totals), np.min(totals), np.max(totals)
```

```python
env.seed(42)

frames = []

obs = env.reset()
for step in range(200):
    img = env.render(mode="rgb_array")
    frames.append(img)
    action = basic_policy(obs)

    obs, reward, done, info = env.step(action)
    if done:
        break
```

```python
def update_scene(num, frames, patch):
    patch.set_data(frames[num])
    return patch,

import matplotlib.animation as animation
mpl.rc('animation', html='jshtml')

def plot_animation(frames, repeat=False, interval=40):
    fig = plt.figure()
    patch = plt.imshow(frames[0])
    plt.axis('off')
    anim = animation.FuncAnimation(
        fig, update_scene, fargs=(frames, patch),
        frames=len(frames), repeat=repeat, interval=interval)
    plt.close()
    return anim
```

```python
plot_animation(frames)
# 시간이 지날수록 카트가 점점 더 크게 진동하여 막대가 심하게 기울어지게 된다.
```
